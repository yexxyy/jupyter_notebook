```python
import numpy as np
def format_float(x, precision, mode='d'):
    """
    :param x:
    :param precision: 小数点位数
    :param mode: d=default, c=ceil, f=floor
    :return:
    """
    scale = 10 ** precision
    if mode == 'c':
        return np.ceil(x * scale) / scale
    elif mode == 'f':
        return np.floor(x * scale) / scale
    else:
        return float(np.format_float_positional(x, precision=int(precision)))
format_float(0, 0, 'c')
```
