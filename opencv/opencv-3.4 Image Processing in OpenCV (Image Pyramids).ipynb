cv2.pyrUp(), cv2.pyrDown()

### Theory

There are two kinds of Image Pyramids.

1) Gaussian Pyramid and 
Higher level (Low resolution) in a Gaussian Pyramid is formed by removing consecutive rows and columns in Lower level (higher resolution) image. Then each pixel in higher level is formed by the contribution from 5 pixels in underlying level with gaussian weights.

2) Laplacian Pyramids

Laplacian Pyramids are formed from the Gaussian Pyramids. There is no exclusive function for that. Laplacian pyramid images are like edge images only.
