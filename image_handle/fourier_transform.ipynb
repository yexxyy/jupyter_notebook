```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from scipy import signal,misc
```

```python
def show(img):
    fig,ax=plt.subplots()
    plt.imshow(img)
    plt.gray()
    plt.show()

img=misc.face(gray=True)
show(img)
```

```python
img
```

```python
f=np.fft.fft2(img)
f_shift=np.fft.fftshift(f)
fimg=np.log(np.abs(f_shift))
fimged=fimg/fimg.max()
show(fimged)
fimged.max(),fimged.min()
```

```python
import cv2

mask=np.ones(fimged.shape)
mask=cv2.circle(mask,(int(fimged.shape[1]/2),int(fimged.shape[0]/2)),100,(0,0,0),-1)
show(mask)
```

```python
masked=np.where(mask==1,fimged,0)
show(masked)
```
